```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
file = os.path.join(os.getcwd(), 'final_data_analysis', 'sample_generation_data.csv')
data = pd.read_csv(file)
```

```python
plot_dir = os.path.join(os.getcwd(), 'final_plots', 'timing')
```

```python
data["time"] = data["time"].str.strip(" ms").astype("float")
```

```python
binwidth=5
fig, ax = plt.subplots()
n, bins, patches = plt.hist(data["time"], bins=range(25, 110, binwidth), edgecolor="black")
ax.bar_label(patches)
plt.title("Synthetic data generation time (n=1000)")
plt.ylabel("Count")
plt.xlabel("Time (ms)")

plt.figtext(0.67,0.8, 
            f"Average: {round(data["time"].mean(), 2)} ms\nSD: {round(data["time"].std(), 2)} ms",
            wrap = True, fontsize = 10, 
            bbox ={'facecolor':'grey', 
                   'alpha':0.3, 'pad':5})

fig.savefig(os.path.join(plot_dir,  "sample_gen_time.png"), dpi=300, bbox_inches="tight")
```
